```python
'''
匯入套件
'''
# 操作 browser 的 API
from selenium import webdriver

# 處理逾時例外的工具
from selenium.common.exceptions import TimeoutException

# 面對動態網頁，等待某個元素出現的工具，通常與 exptected_conditions 搭配
from selenium.webdriver.support.ui import WebDriverWait

# 搭配 WebDriverWait 使用，對元素狀態的一種期待條件，若條件發生，則等待結束，往下一行執行
from selenium.webdriver.support import expected_conditions as EC

# 期待元素出現要透過什麼方式指定，通常與 EC、WebDriverWait 一起使用
from selenium.webdriver.common.by import By

# 強制等待 (執行期間休息一下)
from time import sleep

# 整理 json 使用的工具
import json

# 執行 command 的時候用的
import os

options = webdriver.ChromeOptions()
# options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--incognito")               #開啟無痕模式
options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

# 使用 Chrome 的 WebDriver (含 options)
driver = webdriver.Chrome( options = options )
listData = []

'''
以 function 名稱，作為爬蟲流程
'''

# 走訪頁面
def visit():
    driver.get('https://www.104.com.tw/jobs/main/');

# 輸入關鍵字
def search():
    # 輸入名稱
    txtInput = driver.find_element(By.CSS_SELECTOR, "input#ikeyword")
    txtInput.send_keys("python")
    
    sleep(2)
    
    btnInput = driver.find_element(By.CSS_SELECTOR, "span#icity")
    btnInput.click()
    
    sleep(2)
    
    btnInput = driver.find_element(By.CSS_SELECTOR, "div.second-floor-rect input[value='6001001000']")
    btnInput.click()
    btnInput = driver.find_element(By.CSS_SELECTOR, "div.second-floor-rect input[value='6001002000']")
    btnInput.click()
    
    sleep(2)
    
    btnInput = driver.find_element(By.CSS_SELECTOR, "button.category-picker-btn-primary")
    btnInput.click()
    
    sleep(2)
    
    btnInput = driver.find_element(By.CSS_SELECTOR, "button.btn.btn-primary.js-formCheck")
    btnInput.click()

    sleep(2)
    
    btnInput = driver.find_element(By.CSS_SELECTOR, "ul#js-job-tab > li[data-value='1']")
    btnInput.click()

def scroll():
    # 瀏覽器內部的高度
    innerHeightOfWindow = 0
    
    # 當前捲動的量(高度)
    totalOffset = 0
    
    # 在捲動到沒有元素動態產生前，持續捲動
    while totalOffset <= innerHeightOfWindow:
        # 每次移動高度
        totalOffset += 500;
        
        # 捲動的 js code
        js_scroll = '''(
            function (){{
                window.scrollTo({{
                    top:{}, 
                    behavior: 'smooth' 
                }});
            }})();'''.format(totalOffset)
        
        # 執行 js code
        driver.execute_script(js_scroll)
        
        # 強制等待
        sleep(1)
        
        # 透過執行 js 語法來取得捲動後的當前總高度
        innerHeightOfWindow = driver.execute_script(
            'return window.document.documentElement.scrollHeight;'
        );
        
        # 強制等待
        sleep(1)
        
        # 印出捲動距離
        print("innerHeightOfWindow: {}, totalOffset: {}".format(innerHeightOfWindow, totalOffset))
        
        if totalOffset >= 1000:
            break
        
            
            
# 分析頁面元素資訊
def parse():
    # 取得主要元素的集合
    python = driver.find_elements(
        By.CSS_SELECTOR, 
        'div#js-job-content article')
    
    # 逐一檢視元素
    for element in python:
        # 印出分隔文字
        print("=" * 30)    
    # 取得資料名稱
        a = element.find_element(By.CSS_SELECTOR,'div.b-block__left')
        
        b = element.find_element(By.CSS_SELECTOR,'h2.b-tit a.js-job-link')
        btit = b.text
        print(btit)
        
        c = element.find_element(By.CSS_SELECTOR,'h2.b-tit a.js-job-link')
        cLink = c.get_attribute('href')
        print(cLink)
        
        d = element.find_element(By.CSS_SELECTOR,'ul.b-list-inline.b-clearfix.job-list-intro.b-content li')
        dLocation = d.text
        print(dLocation)
        
        e = element.find_element(By.CSS_SELECTOR,'ul.b-list-inline.b-clearfix li a')
        eName = e.get_attribute('title')
        print(eName)
        
        f = element.find_element(By.CSS_SELECTOR,'ul.b-list-inline.b-clearfix li a')
        fcompanyLink = f.get_attribute('href')
        print(fcompanyLink)
        
        # 放資料到 list 中
        listData.append({
            "btit": btit,
            "cLink": cLink,
            "dLocation": dLocation,
            "eName": eName,
            "fcompanyLink": fcompanyLink
        })
        
# 將 list 存成 json
def saveJson():
    fp = open("python.json", "w", encoding='utf-8')
    fp.write( json.dumps(listData, ensure_ascii=False) )
    fp.close()
    
# 關閉瀏覽器
def close():
    driver.quit()
    
def download():
    # 開啟 json 檔案
    fp = open("python.json", "r", encoding='utf-8')
    
    #取得 json 字串
    strJson = fp.read()
    
    # 關閉檔案
    fp.close()
    
    # 將 json 轉成 list (裡面是 dict 集合)
    listResult = json.loads(strJson)
    
if __name__ == '__main__':
    visit()
    search()
    scroll()
    parse()
    saveJson()
    close()
    download()
```
